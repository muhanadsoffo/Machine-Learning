```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("polynomial-regression.csv",sep = ";")

x = df.car_max_speed.values.reshape(-1,1)#extracting the max speed column and shape it into 2D array and save it in x
y = df.car_price.values.reshape(-1,1)# extracting the price values and reshape it into 2D array and save it in y


#process of making the graph and printing it 
plt.scatter(x,y)
plt.xlabel("car_max_speed")
plt.ylabel("car_price")
plt.show()
```

```python
# linear regression =  y = b0 + b1*x
# multiple linear regression   y = b0 + b1*x1 + b2*x2

# %% linear regression
from sklearn.linear_model import LinearRegression

lr = LinearRegression() #creating instance from LinearRegression class

lr.fit(x,y) #fiting the features into x and output into y

#%% predict
y_head = lr.predict(x)
#process of creating the graph and printing it 
plt.scatter(x,y)
plt.xlabel("car_max_speed")
plt.ylabel("car_price")

plt.plot(x,y_head,color="yellow",label ="linear")
plt.show()

print("10 million TL car speed estimation: ",lr.predict([[10000]]))
```

```python
# %%
# polynomial regression =  y = b0 + b1*x +b2*x^2 + b3*x^3 + ... + bn*x^n

from sklearn.preprocessing import PolynomialFeatures
polynomial_regression = PolynomialFeatures(degree = 2) #creating an instance form PolynomialFeatures class and give it a degree of 2

x_polynomial = polynomial_regression.fit_transform(x)#transforming the x into polynomial feature and save it in x_polynomial


# %% fit
linear_regression2 = LinearRegression() #creating an instance from LinearRegression class an assign it to linear_regression2
linear_regression2.fit(x_polynomial,y) #fitting the features and the output

# %%

#process of creating and printing the praph
y_head2 = linear_regression2.predict(x_polynomial)

plt.scatter(x,y)
plt.xlabel("car_max_speed")
plt.ylabel("car_price")
plt.plot(x,y_head,color="red",label ="linear")

plt.plot(x,y_head2,color= "green",label = "poly")
plt.legend()
plt.show()

```

```python
#%%
from sklearn.metrics import r2_score #importing r2_score function to calculate the r_square score for linear and polynomial regressions of predicted and actual values

print("r_square score for linear regression: ", r2_score(y,y_head))

print("r_square score for polynomial regression: ", r2_score(y,y_head2))
```
